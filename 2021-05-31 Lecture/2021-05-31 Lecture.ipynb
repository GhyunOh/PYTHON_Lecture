```python
import matplotlib.cm as cm
N=256
angle = np.linspace(0,8*2*np.pi,N)
radius = np.linspace(5,1,N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X,Y,c=angle,cmap=cm.Spectral) # cm.Greys, cm.hsv
plt.show()
```

```python
N = 4
A = np.array([1,2,3,2])
B = np.array([2,3,2,1])
C = np.array([1,4,2,3])
X = np.arange(N)
plt.bar(X,A,color="r",hatch='/')
plt.bar(X,A+B,color="b",hatch='x',bottom=A)
plt.bar(X,A+B+C,hatch='/',bottom=A+B,color='g')
plt.show()
```

```python
M = np.random.standard_normal((500,2))
R = np.sum(M**2, axis=1)
plt.figure(figsize=(5,5))
plt.ylim(-5,5)
plt.xlim(-5,5)
plt.scatter(M[:,0],M[:,1],c="0.7",marker='s',
           s=32*R, edgecolor='k')
plt.show()
```

```python
X = np.linspace(-6,6,1024)
X_small = np.linspace(-3,3,1024)
Y_small = np.sinc(X_small)

plt.plot(X,np.sinc(X),c='0.7')
sub_axes = plt.axes([0.6,0.6,0.25,0.25])
sub_axes.plot(X_small,Y_small,c="0.0")
plt.show()
```

```python
import matplotlib.patches as patches
plt.axes(polar=True)
radius = [1,1,1,1,1,1,1,1]
theta = np.linspace(0,2*np.pi,len(radius),endpoint=False)
points = np.vstack((theta,radius)).T
plt.gca().add_patch(patches.Polygon(points,
                                    color="0.5"))
plt.show()
```
