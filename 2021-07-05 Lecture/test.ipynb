```python
import pandas as pd
import numpy as np
data = pd.read_excel('./Dry_Bean_Dataset/Dry_Bean_Dataset.xlsx')
```

```python
a = list(dict.fromkeys(data["Class"]))
for j in range(len(a)):
    data[(data['Class'] == a[j])] = j
```

```python
y = data['Class']
X = data.drop(['Class'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=0, stratify=y)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train, y_train)
print("훈련 세트:{:.2f}".format(lasso.score(X_train,y_train)))
print("테스트 세트:{:.2f}".format(lasso.score(X_test,y_test)))
print("사용한 특성의 개수:", np.sum(lasso.coef_ != 0))
```

```python
lasso_check = [0.2]
```

```python
for i in lasso_check:
    lasso = Lasso(alpha=i,max_iter=500000000).fit(X_train,y_train)
    print(i,"훈련 세트 점수: {:.2f}".format(lasso.score(X_train,y_train)))
    print(i,"테스트세트 점수: {:.2f}".format(lasso.score(X_test,y_test)))
    print(i,"사용한 특성의 개수:", np.sum(lasso.coef_ != 0))
    print(i, lasso.coef_)
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
forest = RandomForestClassifier(n_estimators=100)
forest.fit(X_train,y_train)
```

```python
predicted = forest.predict(X_test)
accuracy = accuracy_score(y_test,predicted)
print(forest.score(X_train,y_train))
print(forest.score(X_test,y_test))
print((predicted ==y_test).sum())
print((predicted==y_test).shape[0])
```

```python
importance = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],axis=0)
indices = np.argsort(importance)[::-1]

print("Feature ranking:")

for f in range(X.shape[1]):
    print(X.columns[indices][f],importance[indices[f]])
plt.figure(figsize=(12,8))
plt.bar(range(X.shape[1]),importance[indices],color='r',yerr=std[indices],
       align='center')
plt.xticks(range(X.shape[1]), X.columns[indices],rotation=45)
plt.xlim([-1,X.shape[1]])
plt.show()
```
