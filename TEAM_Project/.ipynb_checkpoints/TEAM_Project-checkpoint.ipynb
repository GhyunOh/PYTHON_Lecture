```python
import pandas as pd
```

```python
df = pd.read_csv("crime.csv", header=[0,1,2],index_col = 0)
df
```

```python
df_reverse = df.transpose()
df_reverse
```

```python
df_reverse = df_reverse.fillna(0)
df_reverse
```

```python
df = df_reverse.assign(강력범죄강간·강제추행 = lambda x: x['강력범죄강간·강제추행']+x['강력범죄유사강간']+x['강력범죄강간']+x['강력범죄강제추행']+
                      x['강력범죄기타강간·강제추행등'])

df.drop(['강력범죄유사강간','강력범죄강간','강력범죄강제추행','강력범죄기타강간·강제추행등'],axis=1,inplace=True)
```

```python
df.head(32)
```

```python
df.loc(axis=0)[:,'M']
```

```python
df_test = df.reset_index()
df_test
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
import seaborn as sn

df_test_cor = df_test.corr()
sn.heatmap(df_test_cor,annot=True)
```

```python
check = df_test['성별'] == 'M'
df_set = df_test[check]
df_set
```

```python
df_set = df_set.drop(['연도','성별','연령'],axis=1)
```

```python
df_set
```

```python
df_man = df_set.sum()
df_man
```

```python
import numpy as np
```

```python
categories = ['강력범죄살인기수','강력범죄살인미수등','강력범죄강도','강력범죄방화','마약범죄']
```

```python
man = df_man[['강력범죄살인기수','강력범죄살인미수등','강력범죄강도','강력범죄방화','마약범죄']]
man = list(man)
man += man[:1]
```

```python
angles = [n/float(len(categories)) * 2 * np.pi for n in range(len(categories))]
angles += angles[:1]
```

```python
plt.figure(figsize=(8,8))
ax=plt.subplot(polar=True)
plt.xticks(angles[:-1], labels=categories, fontsize=13)
ax.set_rlabel_position(30)
ax.plot(angles,man,label=man,linestyle='dashed',color='violet')
ax.fill(angles,man,color='violet',alpha=0.3)
plt.show()
```

```python
check = df_test['성별'] == 'F'
df_set = df_test[check]
df_set = df_set.drop(['연도','성별','연령'],axis=1)
df_female = df_set.sum()

categories = ['강력범죄살인기수','강력범죄살인미수등','강력범죄강도','강력범죄방화']
female = df_female[['강력범죄살인기수','강력범죄살인미수등','강력범죄강도','강력범죄방화']]
female = list(female)
female += female[:1]
angles = [n/float(len(categories)) * 2 * np.pi for n in range(len(categories))]
angles += angles[:1]
plt.figure(figsize=(8,8))
ax=plt.subplot(polar=True)
plt.xticks(angles[:-1], labels=categories, fontsize=13)
ax.set_rlabel_position(30)
ax.plot(angles,female,label=female,linestyle='dashed',color='violet')
ax.fill(angles,female,color='violet',alpha=0.3)
plt.show()
```
